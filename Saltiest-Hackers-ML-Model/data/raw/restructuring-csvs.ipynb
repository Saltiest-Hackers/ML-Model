```python
def valid_comments(df):
    # make note of the original size
    sum_stats["orig_len"].append(df.shape[0])
    
    # slim down to relevant columns first
    df = df[['id', 'by', 'text', 'type']]

    # we only want comments    
    df = df.loc[df['type'] == 'comment']
    df = df.drop(columns='type')

    # and make sure they have text
    df = df.loc[df['text'].isnull() == False]
    
    sum_stats["final_len"].append(df.shape[0])
    
    return df
```

```python
DIR = 'BigQuery-public/'

sum_stats = {
    "orig_len": [],
    "final_len": []   
}

files = [str(i) for i in list(range(1,18))]
count_new = 1

for file in files:
    print('\r', f'Now processing file: {file}', end=' ')
    path = DIR + file
    tmp = pd.read_csv(path)
    tmp = valid_comments(tmp)
    split = np.array_split(tmp, 6)
    print('writing to disk:', end=' ')
    
    for df in split:
        print(f'{df}..', end=' ')
        df.to_csv(f'hn-comments-full/hn{count_new}.csv')
        count_new += 1
```
